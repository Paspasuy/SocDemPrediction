```python
import matplotlib.pyplot as plt
import pandas as pd
from wordcloud import WordCloud
import nltk

from utils import preprocess_text

nltk.download('stopwords')

from nltk.corpus import stopwords

stops = set(stopwords.words('russian'))
print(stops)
```

```python
path = "../../data/"
video_start = pd.read_csv(path + 'video_info_v2.csv')
all_titles = ' '.join(video_start['title'])

print(all_titles[:1000])
```

```python
from utils.text_cleaner import preprocess_text

text = preprocess_text(all_titles)
```

```python
# Функция для визуализации облака слов
def plot_cloud(wordcloud):
    # Устанавливаем размер картинки
    plt.figure(figsize=(40, 30))
    # Показать изображение
    plt.imshow(wordcloud)
    # Без подписей на осях
    plt.axis("off")
```

```python

wordcloud = WordCloud(width = 2000,
                      height = 1500,
                      random_state=1,
                      background_color='black',
                      margin=20,
                      colormap='Pastel1',
                      collocations=False,
                      stopwords = stops).generate(text)
# Рисуем картинку
plot_cloud(wordcloud)
```
